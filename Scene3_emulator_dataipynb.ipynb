```python
if pd.isnull(dg['FirstModDate']).any():
    # Handle or fill the missing values if necessary
    dg['FirstModDate'].fillna('fill_value', inplace=True)  # Replace 'fill_value' with an appropriate value
# Convert the 'EarliestModDate' column to datetime
dg['FirstModDate'] = pd.to_datetime(dg['FirstModDate'], errors='coerce')
initial_year = "2017-12-31"
dataTrain = dg[dg['FirstModDate'] <= initial_year]
dataTest = dg[(dg['FirstModDate'] > initial_year) & (dg['FirstModDate'] <= '2019-11-30')]
validateData = dg[dg['FirstModDate'] > '2019-11-30']
print("Train shape:", dataTrain.shape)
print("Test shape:", dataTest.shape)
print("Validate shape:", validateData.shape)
```

```python
from gensim.models import Word2Vec

def feature_extraction(X_dataTrain, X_dataTest):
    # initialize X_train and X_test
    X_train = []
    X_test = []
    # save models used
    models = []
    # iterate over each column of X_dataTrain
    for i in range(X_dataTrain.shape[1]):
        # check column data type
        if np.issubdtype(X_dataTrain.dtype, np.object):
            # train Word2Vec model using column i
            # print("Training column {}...".format(i))
            # get train and test data
            train_data = np.asarray(X_dataTrain[:, i], dtype=str)  # Convert to numpy array
            test_data = np.asarray(X_dataTest[:, i], dtype=str)  # Convert to numpy array
            # initialize and train Word2Vec model
            w2v = Word2Vec(sentences=train_data, vector_size=100, window=5, min_count=1)
            # transform train and test texts to word embeddings
            train_w2v = [w2v.wv[word] if word in w2v.wv else np.zeros(100) for word in train_data]
            test_w2v = [w2v.wv[word] if word in w2v.wv else np.zeros(100) for word in test_data]
            # if first execution, save only features
            if len(X_train) == 0:
                X_train = np.array(train_w2v)
                X_test = np.array(test_w2v)
            # concatenate existing features
            else:
                X_train = np.concatenate((X_train, np.array(train_w2v)), axis=1)
                X_test = np.concatenate((X_test, np.array(test_w2v)), axis=1)
            # save model
            models.append(w2v)
        else:
            # directly concatenate the column as feature
            if len(X_train) == 0:
                X_train = np.array(X_dataTrain[:, i])
                X_test = np.array(X_dataTest[:, i])
            else:
                X_train = np.concatenate((X_train, np.array(X_dataTrain[:, i])), axis=1)
                X_test = np.concatenate((X_test, np.array(X_dataTest[:, i])), axis=1)
    return X_train, X_test
```

```python
def normalization(X_train,X_test):
    from sklearn.preprocessing import MinMaxScaler
    scaler = MinMaxScaler()
    scaler.fit(X_train)
    X_train_normalized = scaler.transform(X_train)
    X_test_normalized = scaler.transform(X_test)
    return X_train_normalized,X_test_normalized
```

```python

from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
#from sklearn.ensemble import RandomForestClassifier as Classifier
#from skmultiflow.meta.adaptive_random_forests import AdaptiveRandomForest as Classifier
# group by year-month
from sklearn.linear_model import SGDClassifier as Classifier
dataTestTime = dataTest.groupby([dataTest.FirstModDate.dt.year, dataTest.FirstModDate.dt.month])
# get data train labels
y_train = np.array(dataTrain["Malware"])
# remove unused columns from dataTrain
UNUSED_COLUMNS = ["Malware", "sha256", "FirstModDate"]
for c in UNUSED_COLUMNS:
    del dataTrain[c]
# get dataTrain values
X_dataTrain = dataTrain.values
print(X_dataTrain)
# arrays to save information
accs = [] # accuracy
f1s = [] # f1score
recall = [] # recall
precision = [] # precision
years = []
months = []
# iterate over year and month
for (year, month), dataTestMonth in dataTestTime:
    # get data test labels
    y_test = np.array(dataTestMonth["Malware"])
    # remove unused columns from data test
    dataTestMonth = dataTestMonth.drop(columns=UNUSED_COLUMNS)
    # get dataTestMonth values
    X_dataTest = dataTestMonth.values
    # feature extraction
    X_train, X_test = feature_extraction(X_dataTrain, X_dataTest)
    # normalization
    XX_train, XX_test = normalization(X_train, X_test)
    #print("xxtrain",XX_train)
    #print("xxtest",XX_test.shape)
    # classifier
    # clf = RandomForestClassifier(random_state=0)
    clf = Classifier(loss='log', random_state=0)
    # train classifier
    clf.fit(X_train,y_train)
    # predict test data
    # predict test data
    y_pred_proba = clf.predict_proba(X_test)
    y_pred = [np.argmax(pred) for pred in y_pred_proba]
    print(y_pred)
    # update train data and labels with actual month
    X_dataTrain = np.append(X_dataTrain, X_dataTest, axis=0)
    y_train = np.concatenate((y_train, np.array(y_test)), axis=0)
    print("{}-{}".format(year,month))
    # save metrics
    f1 = f1_score(y_test, y_pred)
    acc = accuracy_score(y_test, y_pred)
    re = recall_score(y_test, y_pred)
    pr = precision_score(y_test, y_pred)
    if (f1 !=0) and (re != 0) and (pr != 0):
        accs.append(acc)
        f1s.append(f1)
        recall.append(re)
        precision.append(pr)
        years.append(year)
        months.append(month)
        print(acc)
```
