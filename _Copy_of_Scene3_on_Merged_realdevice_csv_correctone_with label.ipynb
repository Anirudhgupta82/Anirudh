```python
import pandas as pd
import numpy as np

initial_year = "2017-12-31"
dataTrain = dg[dg['EarliestModDate'] <= initial_year]
dataTest = dg[(dg['EarliestModDate'] > initial_year) & (dg['EarliestModDate'] <= '2019-11-30')]
validateData = dg[dg['EarliestModDate'] > '2019-11-30']
print("Train shape:", dataTrain.shape)
print("Test shape:", dataTest.shape)
print("Validate shape:", validateData.shape)
```

```python
from gensim.models import Word2Vec

def feature_extraction(X_dataTrain, X_dataTest):
    # initialize X_train and X_test
    X_train = []
    X_test = []
    # save models used
    models = []
    # iterate over each column of X_dataTrain
    for i in range(X_dataTrain.shape[1]):
        # check column data type
        if np.issubdtype(X_dataTrain.dtype, np.object):
            # train Word2Vec model using column i
            # print("Training column {}...".format(i))
            # get train and test data
            train_data = np.asarray(X_dataTrain[:, i], dtype=str)  # Convert to numpy array
            test_data = np.asarray(X_dataTest[:, i], dtype=str)  # Convert to numpy array
            # initialize and train Word2Vec model
            w2v = Word2Vec(sentences=train_data, vector_size=100, window=5, min_count=1)
            # transform train and test texts to word embeddings
            train_w2v = [w2v.wv[word] if word in w2v.wv else np.zeros(100) for word in train_data]
            test_w2v = [w2v.wv[word] if word in w2v.wv else np.zeros(100) for word in test_data]
            # if first execution, save only features
            if len(X_train) == 0:
                X_train = np.array(train_w2v)
                X_test = np.array(test_w2v)
            # concatenate existing features
            else:
                X_train = np.concatenate((X_train, np.array(train_w2v)), axis=1)
                X_test = np.concatenate((X_test, np.array(test_w2v)), axis=1)
            # save model
            models.append(w2v)
        else:
            # directly concatenate the column as feature
            if len(X_train) == 0:
                X_train = np.array(X_dataTrain[:, i])
                X_test = np.array(X_dataTest[:, i])
            else:
                X_train = np.concatenate((X_train, np.array(X_dataTrain[:, i])), axis=1)
                X_test = np.concatenate((X_test, np.array(X_dataTest[:, i])), axis=1)
    return X_train, X_test
```

```python
# Normalization method:
def normalization(X_train, X_test):
    from sklearn.preprocessing import MinMaxScaler
    scaler = MinMaxScaler()
    scaler.fit(np.asarray(X_train))
    XX_train = scaler.transform(np.asarray(X_train))
    XX_test = scaler.transform(np.asarray(X_test))
    return XX_train, XX_test
# Iterate over months in test data:
```

```python
def normalization(X_train, X_test):
    from sklearn.preprocessing import MinMaxScaler
    from sklearn.impute import SimpleImputer

    # Impute missing values with 0
    imputer = SimpleImputer(strategy='constant', fill_value=0)
    X_train = imputer.fit_transform(X_train)
    X_test = imputer.transform(X_test)

    # Normalize the data
    scaler = MinMaxScaler()
    scaler.fit(X_train)
    X_train_normalized = scaler.transform(X_train)
    X_test_normalized = scaler.transform(X_test)

    return X_train_normalized, X_test_normalized
```
